# **Email Spam Detection Using Perceptron**

NumPy is used for numerical operations such as vector multiplication.

```python
import numpy as np
```

Define the Activation Function. Converts the perceptron’s output into a binary decision (Spam or Not Spam).

```python
def activation(x):
    return 1 if x >= 0 else 0
```

Define the Perceptron Training Function

Trains the perceptron by updating weights and bias using the perceptron learning rule

```python
def train_perceptron(X, y, learning_rate=0.01, epochs=20):
    # Initialize weights and bias
    weights = np.zeros(X.shape[1])
    bias = 0

    # Loop through the dataset multiple times
    for _ in range(epochs):
        for i in range(len(X)):
            linear_output = np.dot(X[i], weights) + bias
            y_pred = activation(linear_output)

            # Calculate error
            error = y[i] - y_pred

            # Update weights and bias
            weights = weights + learning_rate * error * X[i]
            bias = bias + learning_rate * error

    return weights, bias
```

Define the Prediction Function. Uses trained weights and bias to predict whether new emails are spam or not.

```python
def predict(X, weights, bias):
    predictions = []
    for x in X:
        linear_output = np.dot(x, weights) + bias
        predictions.append(activation(linear_output))
    return predictions
```

Create the Training Dataset. Provides sample email data using numerical features for training the perceptron.

```python
# Features: [Number of Suspicious Keywords, Number of Links]
X = np.array([
    [1, 0],   # Not spam
    [3, 1],   # Spam
    [0, 0],   # Not spam
    [4, 2],   # Spam
    [2, 0],   # Not spam
    [5, 3]    # Spam
])

# Labels: 1 = Spam, 0 = Not Spam
y = np.array([0, 1, 0, 1, 0, 1])
```

Train the Perceptron Model. Learns optimal weights and bias from the training data.

```python
weights, bias = train_perceptron(X, y, learning_rate=0.1, epochs=20)

print("Trained Weights:", weights)
print("Trained Bias:", bias)
```

```python
labels = {1: "Spam", 0: "Not Spam"}
```

Test the Model with New Emails. Evaluates how well the perceptron classifies unseen email data.

```python
test_emails = np.array([
    [1, 1],   # Few keywords, one link
    [4, 2],   # Many keywords, many links
    [0, 0]    # Clean email
])

predictions = predict(test_emails, weights, bias)
for i in range(len(test_emails)):
    print(f"Email {i+1}: {labels[predictions[i]]}")
print("Predictions:", predictions)
```

# **# Perceptron for Loan Approval Prediction**

```python
import numpy as np
```

```python
def activation(x):
    if x >= 0:
        return 1
    else:
        return 0
```

```python
def train_perceptron(X, y, learning_rate=0.01, epochs=20):
    # Initialize weights and bias
    weights = np.zeros(X.shape[1])
    bias = 0

    # Training loop
    for _ in range(epochs):
        for i in range(len(X)):
            linear_output = np.dot(X[i], weights) + bias
            y_pred = activation(linear_output)

            # Error calculation
            error = y[i] - y_pred

            # Update weights and bias
            weights = weights + learning_rate * error * X[i]
            bias = bias + learning_rate * error

    return weights, bias
```

```python
def predict(X, weights, bias):
    predictions = []
    for x in X:
        linear_output = np.dot(x, weights) + bias
        predictions.append(activation(linear_output))
    return predictions
```

```python
# Inputs: [Monthly Income, Credit Score]
X = np.array([
    [3000, 650],
    [1500, 500],
    [4000, 700],
    [2000, 550]
])

# Output: 1 = Approved, 0 = Rejected
y = np.array([1, 0, 1, 0])
```

```python
weights, bias = train_perceptron(X, y, learning_rate=0.01, epochs=20)

print("Trained Weights:", weights)
print("Trained Bias:", bias)
```

```python
LoanLabel = {1: "Loan Approved Class", 0: "Loan Rejected"}
```

```python
test_data = np.array([
    [2500, 600],
    [4500, 750]
])

predictions = predict(test_data, weights, bias)

for i in range(len(test_data)):
    print(f"testcase {i+1} → {LoanLabel[predictions[i]]}")

```
