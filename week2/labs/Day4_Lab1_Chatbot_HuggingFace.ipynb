# Week 2, Day 4 -- Lab 1: Build a Chatbot

You'll turn a ready-made **language model** into your own chatbot -- give it a
**personality**, and see how two dials, **temperature** and **top_p**, change how
it talks. No training needed: the model already knows how to chat.

## 2) Load a small chat model

```python
from transformers import pipeline

chat = pipeline("text-generation", model="Qwen/Qwen2.5-0.5B-Instruct",
                torch_dtype="auto", device_map="auto")
```

## 3) One helper to talk to it

`system` sets the bot's personality. `temperature` and `top_p` control how
random the replies are. This is the only cell you need -- everything below just
calls `ask(...)`.

```python
def ask(user, system="You are a helpful assistant.", temperature=0.7, top_p=0.9):
    messages = [{"role": "system", "content": system},
                {"role": "user", "content": user}]
    out = chat(messages, max_new_tokens=200, do_sample=True,
               temperature=temperature, top_p=top_p)
    return out[0]["generated_text"][-1]["content"]
```

## 4) Say hello

```python
print(ask("Tell me a fun fact about space."))
```

## 5) Give it a personality

Change the `system` message and the whole vibe changes.

```python
print(ask("Explain what gravity is.", system="You are a grumpy pirate. Answer in pirate slang."))
```

## 6) The temperature dial

**Low** temperature = focused and predictable. **High** = creative and wild. Run
each a couple of times and compare.

```python
print(ask("Give me a name for a pet robot.", temperature=0.2))   # focused
```

```python
print(ask("Give me a name for a pet robot.", temperature=1.3))   # wild
```

## 7) The top_p dial

`top_p` limits the words the model may pick from to the most likely ones. Lower =
safer, higher = more variety.

```python
print(ask("Describe the ocean in one sentence.", top_p=0.5))
```

## 8) Chat with it (optional)

Run this, type messages, and type `quit` to stop.

```python
system = "You are a friendly science tutor for teenagers."
while True:
    user = input("You: ")
    if user.strip().lower() == "quit":
        break
    print("Bot:", ask(user, system=system))
```
