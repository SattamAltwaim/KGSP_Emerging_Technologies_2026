# Week 2, Day 3 -- Lab 1: Object Detection with YOLO

Yesterday a model told us **what** was in a picture. Today we go further:
**object detection** tells us **what** *and* **where** -- it draws a box around
each object.

We won't train anything. We'll use **YOLO**, a model that has *already* been
trained on millions of images. Reusing a ready-made model like this is called
**transfer learning**.

## 1) Install and load a pretrained YOLO

```python
%pip install ultralytics -q
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")   # already trained to recognise 80 everyday objects
```

## 2) Run detection on an image

We just hand the model a picture. It finds the objects and where they are.

```python
results = model("https://ultralytics.com/images/bus.jpg")
```

## 3) Look at the result

Each object gets a labelled **box** around it.

```python
import matplotlib.pyplot as plt

annotated = results[0].plot()        # draw the boxes and labels
plt.figure(figsize=(8, 8))
plt.imshow(annotated[:, :, ::-1])    # show it
plt.axis("off"); plt.show()
```

## 4) What did it find?

```python
for box in results[0].boxes:
    name = model.names[int(box.cls)]
    confidence = float(box.conf)
    print(f"{name}: {confidence:.0%} sure")
```

## Try it yourself

- Swap the link in step 2 for any other image URL and run it again. How many
  objects does YOLO find?
- Notice we never trained the model -- detection is **classification** (*what*)
  plus a **location** (*where*), done by a model someone else already trained.
