# Week 2, Day 3 -- Lab 1: Object Detection with YOLO

Classification tells us **what** is in a picture. **Object detection** goes
further: **what** *and* **where** -- it draws a box around each object.

We won't train anything. We'll use **YOLO**, a model already trained on millions
of images. Reusing a ready-made model like this is **transfer learning**.

## 1) Setup

```python
%pip install ultralytics -q
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt

model = YOLO("yolov8n.pt")   # already trained to recognise 80 everyday objects
```

## 2) A small helper to show results

We'll reuse this every time we run the model.

```python
def show(results):
    annotated = results[0].plot()        # draw the boxes and labels
    plt.figure(figsize=(8, 8))
    plt.imshow(annotated[:, :, ::-1])    # BGR -> RGB
    plt.axis("off"); plt.show()
```

## 3) Detect objects in a sample image

```python
results = model("https://ultralytics.com/images/bus.jpg")
show(results)
```

## 4) What did it find?

```python
for box in results[0].boxes:
    name = model.names[int(box.cls)]
    print(f"{name}: {float(box.conf):.0%} sure")
```

## 5) Try your own image

Run the cell, choose a photo from your computer, and YOLO will detect what's in
it.

```python
from google.colab import files

uploaded = files.upload()
my_image = next(iter(uploaded))
show(model(my_image))
```

## Recap

Detection = **classification** (*what*) + a **location** (*where*), produced by a
model someone else already trained for us.
